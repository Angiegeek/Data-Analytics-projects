#1.¿Qué es la regresión multivariable y cual es su objetivo principal?


La regresión multivariable es una técnica estadística utilizada para analizar la relación entre una variable dependiente (en este caso, el precio de los aguacates) y múltiples variables independientes (factores que pueden afectar el precio de los aguacates). El objetivo principal de la regresión multivariable es modelar y entender cómo estas variables independientes influyen en la variable dependiente, permitiendo predecir el valor de la variable dependiente en función de los valores de las variables independientes.

Al emplear la regresión multivariable, indentificamos qué variables independientes tienen una relación significativa con el precio de los aguacates y cuál es la naturaleza de esta relación (positiva o negativa). Además, podemos utilizar el modelo resultante para predecir el precio de los aguacates en función de los valores de estas variables independientes, lo que podría ser útil para tomar decisiones comerciales, como la fijación de precios, la gestión de inventario y la planificación de la producción.






#2.¿Cuál es la diferencia entre variables independientes y variables dependientes en el contexto de la regresión multivariable?

 La variable dependiente es aquella que se intenta predecir o explicar, mientras que las variables independientes son aquellas que se utilizan para realizar la predicción o explicación. La regresión multivariable analiza cómo las variables independientes influyen en la variable dependiente y cómo se relacionan entre sí.

#3. Menciona al menos tres supuestos clave de la regresión multivariable y  explícalos brevemente.

La regresión multivariable se basa en varios supuestos clave que deben ser considerados al interpretar los resultados del modelo. Aquí hay tres de los supuestos más importantes:

* **Linealidad**: Uno de los supuestos fundamentales de la regresión multivariable es que la relación entre las variables independientes y la variable dependiente es lineal. Esto significa que el efecto de un cambio en una variable independiente sobre la variable dependiente es constante y proporcional a lo largo de todo el rango de valores de las variables involucradas. Si esta suposición no se cumple, los resultados del modelo pueden ser sesgados o inexactos.

* **Homocedasticidad**: Este supuesto implica que la variabilidad de los errores (residuos) del modelo debe ser constante en todos los niveles de las variables independientes. En otras palabras, la dispersión de los errores no debe cambiar a medida que cambian los valores de las variables independientes. Si hay una violación de este supuesto (heterocedasticidad), los errores pueden ser inconsistentes y afectar la precisión de las estimaciones de los coeficientes.

* **Independencia de los errores**: Se asume que los errores del modelo no están correlacionados entre sí. Esto significa que el error asociado con una observación no está influenciado por el error de otra observación. Si hay autocorrelación entre los errores, puede afectar la precisión de las estimaciones de los coeficientes y los intervalos de confianza asociados.

Estos supuestos son fundamentales para el correcto funcionamiento de la regresión multivariable y deben ser evaluados y, cuando sea posible, verificados antes de interpretar los resultados del modelo. En caso de que alguno de estos supuestos no se cumpla, pueden ser necesarias correcciones o ajustes en el análisis.







#4.¿Cómo interpretas el coeficiente de una variable independiente en una regresión multivariable?

En una regresión multivariable, los coeficientes de las variables independientes representan el cambio esperado en la variable dependiente por cada unidad de cambio en la variable independiente, manteniendo constantes todas las demás variables independientes incluidas en el modelo.

Para interpretar el coeficiente de una variable independiente específica en una regresión multivariable, sigue estos pasos:

1. **Identificar la variable**: Primero, asegúrate de identificar qué variable independiente estás interpretando.

2. **Signo del coeficiente**: Observa el signo del coeficiente. Si el coeficiente es positivo, indica que hay una relación positiva entre la variable independiente y la variable dependiente. Por el contrario, si el coeficiente es negativo, sugiere una relación negativa.

3. **Magnitud del coeficiente**: La magnitud del coeficiente indica la fuerza de la relación entre la variable independiente y la variable dependiente. Cuanto mayor sea el valor absoluto del coeficiente, mayor será el efecto de la variable independiente en la variable dependiente.

4. **Unidad de cambio**: Considera la unidad de cambio de la variable independiente. Por ejemplo, si la variable independiente es el país de origen y el coeficiente es 0.05, esto significa que, manteniendo todas las demás variables constantes, se espera un aumento de 0.05 unidades en la variable dependiente por cada unidad de aumento en el precio del país de origen.

5. **Control de otras variables**: Es importante tener en cuenta que la interpretación del coeficiente de una variable independiente se realiza manteniendo constantes todas las demás variables independientes incluidas en el modelo. Esto significa que el efecto de la variable independiente puede cambiar si las otras variables también cambian.

En resumen, el coeficiente de una variable independiente en una regresión multivariable proporciona información sobre la dirección, magnitud y efecto de esa variable en la variable dependiente, considerando el efecto de las otras variables incluidas en el modelo.

#5.¿Cómo cambia la interpretación de un coeficiente cuando se incluyen múltiples variables independientes en el modelo?

Cuando se incluyen múltiples variables independientes en un modelo de regresión multivariable, la interpretación de los coeficientes de cada variable independiente se vuelve más compleja debido a la presencia de otras variables que también pueden influir en la variable dependiente. Aquí hay algunos puntos importantes a considerar sobre cómo cambia la interpretación de un coeficiente cuando se incluyen múltiples variables independientes:

1. **Interpretación condicional**: En un modelo multivariable, el coeficiente de una variable independiente se interpreta condicionalmente a los valores de las otras variables independientes en el modelo. Esto significa que el efecto de una variable independiente en la variable dependiente puede cambiar dependiendo de los valores de las otras variables incluidas en el modelo.

2. **Interacción entre variables**: La inclusión de múltiples variables independientes puede revelar interacciones complejas entre las variables predictoras. Esto significa que el efecto de una variable independiente en la variable dependiente puede depender no solo de su propio valor, sino también de los valores de otras variables en el modelo.

3. **Control de variables adicionales**: Al incluir múltiples variables independientes en el modelo, se tiene la capacidad de controlar los efectos de otras variables que también pueden influir en la variable dependiente. Esto permite una interpretación más precisa de cómo cada variable independiente contribuye al resultado final.

4. **Ajuste del efecto marginal**: En presencia de múltiples variables independientes, es importante ajustar el efecto marginal de cada variable para tener en cuenta los efectos de las otras variables en el modelo. Esto implica interpretar cómo cambia la variable dependiente en respuesta a un cambio en una variable independiente, manteniendo constantes todas las demás variables independientes en el modelo.

En resumen, la inclusión de múltiples variables independientes en un modelo de regresión multivariable añade complejidad a la interpretación de los coeficientes, ya que se deben considerar las interacciones entre variables y los efectos condicionales de cada variable independiente en presencia de otras variables en el modelo.

Aquí hay una explicación de la importancia de los p-values:

1. **Evaluación de la significancia**: Los p-values indican la probabilidad de observar el efecto del coeficiente de regresión (o incluso un efecto más extremo) si la verdadera relación entre las variables independientes y la variable dependiente es cero. En otras palabras, los p-values ayudan a determinar si el efecto observado es probablemente real o simplemente el resultado del azar. Si el p-value es bajo (generalmente menos de 0.05), se considera que el coeficiente es estadísticamente significativo, lo que significa que es poco probable que el coeficiente sea cero en la población.

2. **Toma de decisiones**: Los p-values ayudan en la toma de decisiones sobre qué variables incluir en el modelo. Si el p-value asociado con un coeficiente es alto, esto sugiere que la variable independiente correspondiente no tiene un impacto significativo en la variable dependiente y puede ser eliminada del modelo. Por otro lado, si el p-value es bajo, indica que la variable independiente es relevante y debe mantenerse en el modelo.

3. **Interpretación del modelo**: Los p-values también ayudan en la interpretación del modelo. Si un coeficiente tiene un p-value bajo, significa que la variable independiente asociada tiene un impacto significativo en la variable dependiente. Esto proporciona información sobre las relaciones entre las variables en el contexto del problema que se está investigando.

En resumen, los p-values son una herramienta crucial en la evaluación de la significancia estadística de los coeficientes de regresión, lo que ayuda a tomar decisiones informadas sobre qué variables incluir en el modelo y cómo interpretar los resultados del modelo de regresión.

#8. Menciona al menos dos razones por las cuales podrías querer transformar variables en un modelo de regresión multivariable.


Transformar variables en un modelo de regresión multivariable puede ser beneficioso por varias razones:

1. **Normalización de la distribución**: Las transformaciones pueden ayudar a normalizar la distribución de los datos. En muchos casos, las variables pueden tener distribuciones sesgadas o no normales, lo que puede afectar la precisión y la interpretación del modelo. Transformar estas variables utilizando funciones como la transformación logarítmica, raíz cuadrada o Box-Cox puede hacer que las distribuciones se aproximen más a la normalidad, lo que puede mejorar la estabilidad y la eficacia del modelo.

2. **Linealidad**: En un modelo de regresión lineal, se asume una relación lineal entre las variables independientes y la variable dependiente. Sin embargo, esta relación puede no ser lineal en la práctica. Transformar las variables puede ayudar a capturar mejor la verdadera relación entre las variables al hacer que la relación sea más lineal. Por ejemplo, si la relación entre las variables es logarítmica o cuadrática, aplicar transformaciones logarítmicas o cuadráticas a las variables puede ayudar a modelar esa relación de manera más efectiva.

3. **Homocedasticidad**: La homocedasticidad se refiere a la constancia de la varianza de los errores en un modelo de regresión. Si la varianza de los errores es heterocedástica, es decir, varía en diferentes niveles de las variables independientes, puede afectar la validez de las inferencias estadísticas del modelo. Aplicar transformaciones a las variables puede ayudar a estabilizar la varianza de los errores y mejorar la homocedasticidad del modelo.

4. **Reducción de sesgo o mejorar la relación con la variable dependiente**: Algunas transformaciones pueden ayudar a reducir el sesgo en los datos o mejorar la relación con la variable dependiente. Por ejemplo, si hay una relación no lineal entre una variable independiente y la variable dependiente, aplicar una transformación puede ayudar a capturar mejor esa relación y reducir el sesgo en las predicciones del modelo.

#9.¿Cómo puedes verificar visualmente la normalidad de los residuos después de ajustar un modelo?

Podemos hacer la verificación visual mediante diversos gráficos, a continuación los siguientes ejemplos:

```python
# Realiza predicciones utilizando el modelo ajustado
y_pred = modelo.predict(X)  # Donde X son tus variables independientes

# Calcula los residuos restando las predicciones del modelo de los valores observados
residuos = y - y_pred


# Histograma de residuos
plt.figure(figsize=(8, 6))
plt.hist(residuos, bins=20, density=True, alpha=0.6, color='b')
plt.title('Histograma de Residuos')
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')

plt.show()
```

Histograma de Residuos: Grafica un histograma de los residuos y compáralo con una distribución normal teórica. Si los residuos siguen una distribución normal, el histograma debería tener una forma de campana simétrica alrededor de cero.


```python
# Gráfico de Probabilidad Normal
plt.figure(figsize=(8, 6))
stats.probplot(residuos, dist="norm", plot=plt)
plt.title('Gráfico de Probabilidad Normal de los Residuos')

plt.show()
```

Gráfico de Probabilidad Normal: Este es un gráfico de dispersión en el que los valores ordenados de los residuos se representan en el eje vertical y los valores esperados de los residuos se representan en el eje horizontal. Si los residuos siguen una distribución normal, los puntos en el gráfico deberían formar una línea diagonal.

```python
# Gráfico de Densidad de Kernel
plt.figure(figsize=(8, 6))
sns.kdeplot(residuos, shade=True)
plt.title('Gráfico de Densidad de Kernel de los Residuos')
plt.xlabel('Residuos')
plt.ylabel('Densidad')

plt.show()
```

Gráfico de Densidad de Kernel: Este gráfico muestra la densidad de los residuos y puede ayudar a identificar desviaciones de la normalidad más fácilmente que un histograma.

#10. ¿Cómo afectan los outliers en los datos al modelo de regresión multivariable?


Los outliers, o valores atípicos, pueden afectar de varias maneras a un modelo de regresión multivariable:

1. **Influencia en los coeficientes del modelo**: Los outliers pueden ejercer una influencia desproporcionada en la estimación de los coeficientes del modelo. Debido a que los modelos de regresión lineal minimizan la suma de los cuadrados de los residuos (método de mínimos cuadrados), los outliers pueden tener un impacto significativo en la estimación de los coeficientes, especialmente si están asociados con valores extremadamente altos o bajos de las variables independientes.

2. **Distorsión de la relación entre variables**: Los outliers pueden distorsionar la relación entre las variables independientes y la variable dependiente. Esto puede llevar a una interpretación incorrecta de la relación entre las variables y puede afectar la precisión de las predicciones del modelo.

3. **Afectan la precisión de las predicciones**: Los outliers pueden aumentar el error residual del modelo, lo que afecta la precisión de las predicciones. Esto se debe a que los outliers pueden llevar a una mala generalización del modelo a nuevos datos, ya que el modelo puede sobreajustarse o subajustarse para acomodar estos valores extremos.

4. **Violación de supuestos de normalidad y homocedasticidad**: Los outliers pueden violar los supuestos de normalidad y homocedasticidad de los residuos en un modelo de regresión. Por ejemplo, si un outlier tiene un impacto desproporcionado en la varianza de los residuos, puede llevar a una violación de la homocedasticidad, lo que afecta la validez de las inferencias estadísticas del modelo.

En resumen, los outliers pueden tener un impacto significativo en la precisión y la interpretación de un modelo de regresión multivariable, por lo que es importante identificarlos y considerar estrategias para manejarlos adecuadamente, como la transformación de variables, la eliminación de valores atípicos o el uso de modelos robustos que sean menos sensibles a los outliers.

Estas estrategias pueden ser implementadas en Python utilizando bibliotecas como Pandas, NumPy y Scikit-learn.
```python
#MANEJO DE OUTLIDERS: Rango Intercuantílico (IQR)
# Calcular el rango intercuartílico (IQR) para cada columna
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Definir los límites inferior y superior para considerar valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Alinear los DataFrames para evitar comparaciones incorrectas
df, lower_bound = df.align(lower_bound, axis=1, join='inner')
df, upper_bound = df.align(upper_bound, axis=1, join='inner')

# Filtrar filas que están dentro de los límites establecidos
df_outliers_removed = df[(df >= lower_bound) & (df <= upper_bound)].dropna()

# Mostrar el DataFrame sin outliers
print(df_outliers_removed)

```

Este código implementa una técnica comúnmente utilizada para manejar outliers, conocida como el rango intercuartílico (IQR):

1. **Calcular el rango intercuartílico (IQR)**:
   - Se calculan los cuantiles del 25% (Q1) y del 75% (Q3) para cada columna del DataFrame `df`.
   - El rango intercuartílico (IQR) se calcula como la diferencia entre Q3 y Q1.

2. **Definir los límites inferior y superior para considerar valores atípicos**:
   - Se definen límites inferior y superior multiplicando el IQR por 1.5 y sumando/restando estos valores a Q1 y Q3, respectivamente.
   - Los valores que caen fuera de estos límites se consideran outliers.

3. **Alinear los DataFrames para evitar comparaciones incorrectas**:
   - Antes de filtrar los outliers, los DataFrames `df` y `lower_bound` (límites inferiores) y `df` y `upper_bound` (límites superiores) se alinean para asegurar que tengan las mismas etiquetas de índice y columnas antes de realizar la comparación. Esto evita problemas como la comparación incorrecta de valores.

4. **Filtrar filas que están dentro de los límites establecidos**:
   - Se filtran las filas del DataFrame `df` que están dentro de los límites establecidos por `lower_bound` y `upper_bound`. Esto se hace utilizando la sintaxis de filtrado booleano para seleccionar solo las filas donde cada valor está dentro de los límites.

5. **Mostrar el DataFrame sin outliers**:

#13. ¿Cómo interpretarías los resultados y los coeficientes obtenidos?


1. **MSE (Error Cuadrático Medio)**:
   - El MSE mide la cantidad promedio por la cual las predicciones del modelo difieren de los valores reales al cuadrado.
   - En este caso, un MSE de 0.0973 indica que, en promedio, las predicciones del modelo difieren de los valores reales en aproximadamente 0.0973 unidades al cuadrado.
   - Cuanto menor sea el MSE, mejor será el rendimiento del modelo en términos de precisión de las predicciones.

2. **RMSE (Raíz del Error Cuadrático Medio)**:
   - El RMSE es simplemente la raíz cuadrada del MSE y proporciona una medida de la magnitud del error en la misma escala que la variable dependiente.
   - En este caso, un RMSE de 0.3121 indica que, en promedio, las predicciones del modelo difieren de los valores reales en aproximadamente 0.3121 unidades.
   - Cuanto menor sea el RMSE, mejor será el rendimiento del modelo en términos de precisión de las predicciones.

3. **R2 (Coeficiente de Determinación)**:
   - El coeficiente de determinación (R2) es una medida de la proporción de la variabilidad en la variable dependiente que es explicada por el modelo. Es un valor entre 0 y 1, donde 1 indica un ajuste perfecto.
   - En este caso, un R2 de 0.3963 indica que aproximadamente el 39.63% de la variabilidad en la variable dependiente es explicada por el modelo.
   - Cuanto más cercano a 1 sea el R2, mejor será la capacidad del modelo para explicar la variabilidad en los datos.
